# 1. Extract all the NOUNs from the text

```python
# extract all nouns and append in a list
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(story)

# to store the NOUN tokens
noun_tokens = []
# fetch the pos tags
for token in doc :
    if token.pos_ in ["NOUN"] :
    # print(token, " | ", token.pos_)
        noun_tokens.append(token)
noun_tokens
```
