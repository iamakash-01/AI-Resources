# Removal of HTML tags

```python
import re
def remove_html_tags(text) :
  pattern = re.compile("<.*?>")
  return pattern.sub(r"",text)
```

```python
text = "<html><body><p> Movie 1 </p><p> Actor - Amir Khan</p><p>"
```

```python
remove_html_tags(text)
```

# Applying it to whole dataset
# .apply() takes functions as args

```python
df["review"] = df["review"].apply(remove_html_tags)
```

# Remove URLs

```python
def remove_url(text) :
  pattern = re.compile(r"https?://\S+|www\.\S+")
  return pattern.sub(r"",text)
```

```python
text1 = "Check my github https://github.com"
text2 = "Check my linkedIn https://linkedIn.com"
```

```python
remove_url(text1)
```

# Punctuation Removal

```python
import string
```

```python
string.punctuation
```

```python
exclude = string.punctuation
exclude
```

```python
def punctuation_removal(text) :
  for char in exclude :
    text = text.replace(char," ")
  return text
```

```python
text = "string. Where. Punctuation?"
```

```python
punctuation_removal(text)
```

```python
df["review"][5]
```

```python
df["review"] = df["review"].apply(punctuation_removal)
```

```python
df["review"][5]
```

# STEMMER

```python
# stemmer
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem_words(text) :
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = "Walk Walks Walked Walking"
stem_words(sample)
```
