```python
# module for regular expression
import re
```

```python
chat1 = "codebasics: you ask a lot of questions @@ 1235678912, 0909090909, abc@xyz.com"
chat2 = "codebasics: here it is:(123)-567-8912, abc@xyz.com"
chat3 = "codebasics: yes, phone:1234567912, email:abc@xyz.com"
```

```python
# matching 10 digits for chat1
pattern = "\d{10}"
matches = re.findall(pattern, chat1)
print(matches)
```

```python
# matching with brackets for chat2
pattern = "\(\d{3}\)-\d{3}-\d{4}"
result = re.findall(pattern, chat2)
print(result)
```

```python
# matching with all the sets
pattern = "\d{10}|\(\d{3}\)-\d{3}-\d{4}"
results = re.findall(pattern, chat3)
print(result)
```

```python
# extracting for email ids
pattern = "[a-z0-9A-Z]*@[a-z0-9A-Z]*\.com"
result = re.findall(pattern,chat3)
print(result)
```

```python
# for reteieving order no
text = "codebasics: Hello, I am having an issue with my order 41288912"
pattern = "order[^\d]*(\d*)"
result = re.findall(pattern, text)
print(result)
```

```python
# find the age of elon musk
elon = """Born	Elon Reeve Musk 
June 28, 1971 (age 53)
Pretoria, Transvaal, South Africa
Citizenship	
South Africa
Canada (from 1989)
United States (from 2002)
Education	University of Pennsylvania (BA, BS)
Occupation	Businessman
Title	
Founder, CEO, and chief engineer of SpaceX
CEO and product architect of Tesla, Inc.
Owner, CTO and executive chairman of X (formerly Twitter)
Founder of The Boring Company, X Corp., and xAI
Co-founder of Neuralink, OpenAI, Zip2, and X.com (part of PayPal)
President of the Musk Foundation
Spouses	
Justine Wilson
​
​(m. 2000; div. 2008)​
Talulah Riley
​
​(m. 2010; div. 2012)​
​
​(m. 2013; div. 2016)"""
```

```python
# extract the age
pattern = "age \d+"
res = re.findall(pattern, elon)
print(res)
```

```python
# extract the name
pattern = 'Born(.*)'
res = re.findall(pattern, elon)
print(res)
```

```python
# extract the DOB
pattern = "Born.*\n(.*)\(age"
res = re.findall(pattern,elon)
print(res)
```

```python
# extract the birthplace
pattern = "Born.*\n.*\n(.*)\n"
res = re.findall(pattern,elon)
print(res)
```
