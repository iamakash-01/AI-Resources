```python
# performing lower case operations
df["Text"] = df["Text"].str.lower()
```

```python
# remove html tags
import re
def remove_html_tags(text) :
  pattern = re.compile("<,*?>")
  return pattern.sub(r"",text)
```

```python
df["Text"]
```

```python
# applying in whole dataset
df["Text"] = df["Text"].apply(remove_html_tags)
```

```python
df["Text"][7]
```

```python
# punctuation removals
import string
```

```python
string.punctuation
```

```python
# store the punctuations
exclude = string.punctuation
```

```python
def punctuation_removal(text) :
  for char in exclude :
    text = text.replace(char,"")
  return text
```

```python
punctuation_removal(df["Text"])
```

```python
df["Text"] = df["Text"].apply(punctuation_removal)
```
