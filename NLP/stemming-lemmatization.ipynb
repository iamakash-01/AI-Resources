```python
from nltk.stem import PorterStemmer 
stemmer = PorterStemmer()
```

```python
words = ["eating","eats","ate","adjustable","ability"]
for word in words :
    print(word," | ", stemmer.stem(word))
```

## 2. Lemmatization in spacy

```python
import spacy
# spacy uses a string
nlp = spacy.load("en_core_web_sm")
doc = nlp("eating eats ate adjustable ability")
for token in doc :
    print(token.text," | ",token.lemma_)
```

```python
# attribute ruler
nlp = spacy.load("en_core_web_sm")
ar = nlp.get_pipe("attribute_ruler")
ar.add([[{"TEXT" : "Bro"}],[{"TEXT" : "Bruh"}]],{"LEMMA" : "Brother"})
doc = nlp("Bro, you gonna go? Bruh, don't say no! I am exhausted")
for token in doc :
    print(token.text, " | ", token.lemma_)
```
